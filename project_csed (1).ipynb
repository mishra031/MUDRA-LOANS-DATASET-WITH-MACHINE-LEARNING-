```python
df['Loan_Term'].plot(kind='box')
plt.title('Loan_Term')
plt.show()
```

```python
q1=df['Loan_Term'].quantile(0.25)
q2=df['Loan_Term'].quantile(0.75)
iqr=q2-q1
lower_bound=q1-(1.5*iqr)
upper_bound=q2+(1.5*iqr)
```

```python
filtered_df=df[(df['Loan_Term']>=lower_bound) & (df['Loan_Term']<=upper_bound)]
filtered_df
```

```python
filtered_df['Loan_Term'].plot(kind='box')
plt.title('Loan_Term')
plt.show()

```
